```python
from llama_index.core import SimpleDirectoryReader

# 1. Define the path to your test paper
# We use '../' to go up one level from the 'notebooks' folder into 'data'
file_path = "../data/papers/triples/ToG.pdf"

print(f"Attempting to load: {file_path}")

# 2. Load the PDF using the standard, out-of-the-box parser
documents = SimpleDirectoryReader(input_files=[file_path]).load_data()

print(f"Successfully loaded {len(documents)} pages.\n")

# 3. Inspect the raw text of a complex page
# Let's look at page 4 (index 3), which usually contains math or tables in academic papers.
page_to_inspect = 3 

print(f"--- Raw Text Output from Page {page_to_inspect + 1} --- \n")
print(documents[page_to_inspect].text[:1500]) # Printing the first 1500 characters
```
